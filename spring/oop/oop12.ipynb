# <font color=blue> Разработка через тестирование</font>

<img src="https://miro.medium.com/max/840/1*tZSwCigaTaJdovyWlp5uBQ.jpeg">

1. Написать тест

2. Убедиться, что тест "падает", то есть что система не проходит тест.

3. Максимально быстро написать код для прохождения теста.

4. Убедиться, что тест пройден.

5. Рефакторинг или чистка кода. 

6. Вернуться к пункту 1.


### Три закона TDD

1. Новый рабочий код пишется только для прохождения непройденного теста.

2. Нельзя за раз добавлять более одного теста, который система не проходит.

3. Пишите не больше рабочего кода, чем необходимо для прохождения теста.

## <font color=green>Зачем нужно?</font>

1. Снижение числа ошибок. Согласно исследованиям, число багов к первому выпуску продукта на 40 - 90 % ниже.

2. Помогает работать быстро с большими и сложными проектами:
  
  - чтобы легко справляться с большим проектом, код в нем должен быть **чистым**;
  
  - чтобы код был чистым, нужно его переодически рефакторить;
  
  - чтобы рефакторить код, надо быть уверенным, что в процессе ничего не сломается;
  
  - тесты позволяют мгновенно заметить, что внесенное изменение привело к ошибке.
  
3. Тесты пишутся **до** написания кода, так как программистам свойственно обходить свои ошибки стороной. Таким образом, тесты, написанные к готовому рабочему коду, менее эффективны. Кроме того, написание тестов к уже готовому своему коду - скучное занятие.

4. TDD подталкивает  начинающих разработчиков к написанию, кода, который легко тестировать.

5. TDD может служить инструментом в разработке системы, если нет четкого понимания, как она будет устроена. В циклах TDD постепенно формируется дизайн программы.

6. Тесты служат отличной документацией к проекту. При последовательной разработке через тестирование легко понять, что делает публичный метод, по тестам к нему.

## <font color=green>Недостатки</font>

1. Увеличение времени, затрачиваемого на разработку. Такое увеличение может быть очень серьезным. 

2. При серьезной перестройке системы, затрагивающей тестируемые интерфейсы, тесты также придется менять, что будет крайне трудоемкой задачей.

3. Может создавать у разработчика ложную убежденность в низком количестве ошибок.

### Упражнение 1. Делители
