# Задача 1. Хорошие слова

Назовём слово хорошим, если выполнено одно из условий.

- Оно равно слову «0».

- Оно начинается на символ «1», а затем следует хорошее слово.

- Оно начинается на символ «2», а затем следует два записанных друг за другом хороших слова.

- Оно начинается на символ «3», а затем следует три записанных друг за другом хороших слова.

Примеры хороших слов: `10`, `200`, `3000`, `21010`, `2010`, `3020010`, `11111110`, `30011111110`.

Определите, является ли поданное на вход слово хорошим или плохим.

Решите задачу без рекурсии.

Гарантируется, что во входных данных есть только символы `'0'`, `'1'`, `'2'`, `'3'` и что входная последовательность непустая.

### Критерии

За правильное решение **без рекурсии**, работающее за линейное время начисляются 3 балла.

````python
word = input()

def check_word(word):
    stack = []
    n = len(word)
    for i in range(n):
        m = int(word[i])
        if m == 0:
            while len(stack) > 0:
                stack[-1] -= 1
                if stack[-1] == 0:
                    stack.pop()
                break
            if len(stack) == 0:
                if i == n-1:
                    return True
                return False
        else:
            stack.append(m)
    if len(stack) > 0:
        return False
    assert False
            
print(check_word(word))
````

````python
def check_word_2(word):
    n = len(word)
    if word[0] == '0' and n > 1:
        return False
    word = int(word)
    num_of_good_numbers = 0
    for _ in range(n):
        d = word % 10
        word //= 10
        num_of_good_numbers -= d
        if num_of_good_numbers < 0:
            return False
        num_of_good_numbers += 1
    if num_of_good_numbers == 1:
        return True
    return False


print(check_word_2(input()))
````

### Критерии

За правильное решение начисляются 3 балла. **Все** унаследованные методы классов `BSTWithIndexing` и `BSTWithIndexingNode` должны правильно работать. Будьте внительны.

````python
class BSTNode:
    # Description of __slots__ attribute can be found at
    # https://pythonz.net/references/named/object.__slots__/
    # Here __slots__ is used in __str__() method.
    __slots__ = ('key', 'value', 'left', 'right', 'parent')
    
    def __init__(self, key, value, left=None, right=None, parent=None):
        self.key = key
        self.value = value
        self.left = left
        self.right = right
        self.parent = parent
        
    def __str__(self):
        print_str = []
        for k in self.__slots__:
            if k in ['left', 'right', 'parent']:
                n = getattr(self, k)
                if n is None:
                    v = None
                else:
                    k += '.key'
                    v = n.key
                print_str.append('{}={}'.format(k, repr(v)))
            else:
                print_str.append('{}={}'.format(k, repr(getattr(self, k))))
        return self.__class__.__name__ + '(' + ', '.join(print_str) + ')'
    
    def get_attribute_values_string(self, attributes):
        """Returns a string with names and values of
        specified attributes. Used in Tree.prefix_traverse()
        method for optional demostration of node contents."""
        
        values = [getattr(self, attr) for attr in attributes]
        attr_strings = ['{}={}'.format(attr, repr(v)) for attr, v in zip(attributes, values)]
        return ' '.join(attr_strings)
    
    def _get_slots_attributes(self):
        attributes = {}
        for attr in self.__slots__:
            attributes[attr] = getattr(self, attr)
        return attributes
        
    def copy(self):
        kwargs = self._get_slots_attributes()
        return self.__class__(**kwargs)
        

class BST:
    def __init__(self):
        self.root = None

    ############################
    # Traverse methods
        
    @staticmethod
    def _get_connection_lines_for_children(connections):
        connections = connections.copy()
        if len(connections) == 0:
            return connections
        if connections[-1] == '├':
            connections[-1] = '│'
        elif connections[-1] == '└':
            connections[-1] = ' '
        else:
            raise ValueError('no connection to current node')
        return connections
            
    @staticmethod
    def _print_connections(connections):
        connections_str = ''
        for c in connections:
            if c == ' ':
                connections_str += c*4
            elif c == '│':
                connections_str += c + ' '*3
            elif c in '├└':
                connections_str += c + '─'*2 + ' '
            else:
                raise ValueError("unknown character '{}' in connections list".format(c))
        print(connections_str, end='')
        
    def _prefix_traverse(self, root, connections, attributes):
        self._print_connections(connections)
        connections = self._get_connection_lines_for_children(connections)
        if root is not None:
            print(root.key, root.get_attribute_values_string(attributes))
            connections_right = connections.copy()
            connections_right.append('├')
            self._prefix_traverse(root.right, connections_right, attributes)
            connections_left = connections.copy()
            connections_left.append('└')
            self._prefix_traverse(root.left, connections_left, attributes)
        else:
            print('-')
        
    def prefix_traverse(self, attributes=()):
        """Prefix traverse which draws connections between nodes.
        `attributes` is a tuple containing names of attributes
        of tree node which will be printed during traversing.
        For instance if you need to print values in nodes you
        may use
        ```python
        import random
        for _ in range(5):
            key = random.randint(0, 5)
            bst.insert(key, str(key))
        bst.prefix_traverse(('value',))
        ```
        The output will look similar to
        ```
        8 value='8'
        ├── 12 value='12'
        │   ├── 13 value='13'
        │   │   ├── -
        │   │   └── -
        │   └── 9 value='9'
        │       ├── -
        │       └── -
        └── 3 value='3'
            ├── -
            └── -
        ```
        By default only keys of the node are printed.
        The right subtree is printed before the left subtree.
        If a node has no child, hyphen is printed instead of 
        child's key.
        Args:
            attributes: a tuple of strings
        
        Returns:
            None
        """
        self._prefix_traverse(self.root, [], attributes)
        
    ##################################
    # Insertion methods
        
    def _insert(self, key, value, root):
        if key == root.key:
            root.value = value
        elif key < root.key:
            if root.left is None:
                root.left = BSTNode(key, value, parent=root)
            else:
                self._insert(key, value, root.left)
        else:
            if root.right is None:
                root.right = BSTNode(key, value, parent=root)
            else:
                self._insert(key, value, root.right)
        
    def insert(self, key, value):
        if self.root is None:
            self.root = BSTNode(key, value)
        else:
            self._insert(key, value, self.root)
            
    ###################################
    # Search methods
            
    def _find(self, key, root):
        if root is None:
            return None
        if key == root.key:
            return root  # ATTENTION changed from root.value
        elif key < root.key:
            return self._find(key, root.left)
        else:
            return self._find(key, root.right)
        
    def find(self, key):
        node = self._find(key, self.root)
        if node is None:
            raise KeyError("key {} is not in tree".format(key))
        return node.value
````

````python
class BSTWithIndexingNode(BSTNode):
    __slots__ = BSTNode.__slots__ + ('left_size', 'right_size')
    
    def __init__(
            self,
            key,
            value,
            left=None,
            right=None,
            parent=None,
            left_size=0,
            right_size=0
    ):
        super().__init__(key, value, left, right, parent)
        self.left_size=left_size
        self.right_size=right_size
        
        
class BSTWithIndexing(BST):
    def __init__(self):
        super().__init__()
        self.size = 0
        
    ##################################
    # Insertion methods
        
    def _insert(self, key, value, root):
        if key == root.key:
            root.value = value
            num_added_nodes = 0
        elif key < root.key:
            if root.left is None:
                root.left = BSTWithIndexingNode(key, value, parent=root)
                num_added_nodes = 1
                self.size += 1
            else:
                num_added_nodes = self._insert(key, value, root.left)
            root.left_size += num_added_nodes
        else:
            if root.right is None:
                root.right = BSTWithIndexingNode(key, value, parent=root)
                num_added_nodes = 1
                self.size += 1
            else:
                num_added_nodes = self._insert(key, value, root.right)
            root.right_size += num_added_nodes
        return num_added_nodes
        
    def insert(self, key, value):
        if self.root is None:
            self.root = BSTWithIndexingNode(key, value)
            self.size = 1
        else:
            _ = self._insert(key, value, self.root)
            
    ###################################
    # Search methods
            
    def _find_kth(self, root, k):
        nl = root.left_size
        if k == nl:
            return root
        elif k < nl:
            return self._find_kth(root.left, k)
        else:
            return self._find_kth(root.right, k-nl-1)
            
        
    def __getitem__(self, k):
        if k < self.size:
            return self._find_kth(self.root, k)
        raise IndexError(
            "Index out of range. Can not get {}th element"
            " because tree has {} elements".format(k, self.size)
        )
````

# Задача 3. Сеть компании Bmail

Когда-то давно в небезызвестной компании Bmail был только один маршрутизатор. Шли года и с течением времени закупались новые маршрутизаторы. Каждый раз при покупке нового маршрутизатора он соединялся с одним из купленных до него. Вам заданы значения $p_i$ — номер маршрутизатора к которому был подключен $i$-й после покупки ($p_i < i$).

Сейчас в Bmail всего $n$ маршрутизаторов. Выведите последовательность маршрутизаторов на пути от первого до $n$-го.

### Входные данные
В первой строке записано целое число $n$ ($2 \le n \le 200000$) — количество маршрутизаторов. Далее во второй строке записано $n−1$ целое число $p_2, p_3, \ldots, p_n$ ($1 \le p_i <i$), где $p_i$ равно номеру маршрутизатора, к которому подключили $i$-й после покупки.

### Выходные данные
Выведите путь от $1$-го до $n$-го маршрутизатора. Пусть должен начинаться с числа $1$ и заканчиваться числом $n$. Все номера в пути должны быть различны.

### Примеры работы программы

| <font size=3>Входные данные</font> | <font size=3>Выходные данные</font> |
| :---: | :---: |
| <font size=3>8<br>1 1 2 2 3 2 5</font> | <font size=3>1 2 5 8</font> |
| <font size=3>6<br>1 2 3 4 5</font> | <font size=3>1 2 3 4 5 6</font> |
| <font size=3>7<br>1 1 2 3 4 3</font> | <font size=3>1 3 7</font> |

### Критерии

За правильное решение начисляется 1 балл.

### Страница с проверяющей системой

http://codeforces.com/problemset/problem/1057/A

````python
n = int(input())

connections = list(map(int, input().split()))

path = [n]
i = n - 2
while i >= 0:
    i = connections[i]
    path.append(i)
    i -= 2

m = len(path)
for i in range(m-1, -1, -1):
    print(path[i], end=' ')
````

# Задача 4. Сумма упорядоченных последовательностей

На вход подаются два упорядоченных по возрастанию списка целых чисел `a` и `b`. Распечатайте $k$ наименьших сумм вида `a[i] + b[j]`, упорядоченных по возрастанию. Алгоритм должен работать с эффективностью не хуже $O(k \cdot \log k)$.

Гарантируется, что `len(a) * len(b) >= k`.

Сначала на вход подается число `k`, затем в отдельных строках списки `a` и `b`.  На выходе должна быть строка, в которой через пробел следуют $k$ наименьших сумм.

### Примеры работы программы

| <font size=3>Входные данные</font> | <font size=3>Выходные данные</font> |
| :---: | :---: |
| <font size=3>3<br>1 2 3<br>1 1 1</font> | <font size=3>2 2 2</font> |
| <font size=3>3<br>1 2 3<br>1 2 3</font> | <font size=3>2 3 3</font> |
| <font size=3>3<br>1 1 1<br>1 1 1</font> | <font size=3>2 2 2</font> |
| <font size=3>4<br>1 2<br>1 2</font> | <font size=3>2 3 3 4</font> |

### Критерии

В задаче есть большой простор для оптимизации. За решение, которое я взял в качестве основного, начисляются 3 балла. Вы можете получить меньше или больше баллов в зависимости от того, насколько эффективным будет Ваш код: возможный диапазон 1.5 - 4.5 балла. Асимптотика должна быть **не хуже** $O(k \cdot \log k)$, то есть перебор не подойдет.

````python
from heapq import heappush as push, heappop as pop


def add_sequences(a, b, k):
    len_a = len(a)
    len_b = len(b)
    row_filling = [0] * (k+1)
    col_filling = [0] * (k+1)
    sum_ = []
    heap = []
    push(heap, (a[0]+b[0], 0, 0))
    for _ in range(min(k, len_a*len_b)):
        s, r, c = pop(heap)
        sum_.append(s)
        row_filling[r] += 1
        col_filling[c] += 1
        if r < len_a - 1 and row_filling[r] - row_filling[r+1] == 1:
            push(heap, (a[r+1]+b[c], r+1, c))
        if c < len_b - 1 and col_filling[c] - col_filling[c+1] == 1:
            push(heap, (a[r]+b[c+1], r, c+1))
    return sum_
````

````python
k = int(input())
a = list(map(int, input().split()))
b = list(map(int, input().split()))

print(add_sequences(a, b, k))
````

````python
def add_sequences_all(a, b):
    s = []
    for aa in a:
        for bb in b:
            s.append(aa + bb)
    return sorted(s)
````

````python
import random

k_max = 30

for _ in range(10000):
    k = random.randint(1, k_max)
    an = random.randint(1, k)
    bn = random.randint(k // an + 1, max(k, k // an + 1))
    a = sorted([random.randint(0, 100) for _ in range(an)])
    b = sorted([random.randint(0, 100) for _ in range(bn)])
    correct = add_sequences_all(a, b)[:k]
    tested = add_sequences(a, b, k)
    if correct != tested:
        print('correct:', correct)
        print('tested:', tested)
        break
````
