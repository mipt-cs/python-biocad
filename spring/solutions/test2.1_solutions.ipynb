# Задача 1. Очередь на основе стека

Реализуйте очередь, которая использует 2 стека для своего функционирования: в первый стек элементы помещаются, а из второго извлекаются.

В качестве стеков можно использовать объекты типа `list`, но в таком случае к спискам разрешается применение только методов `list.append()`, `list.pop()` и встроенной функции `len()`.

Очередь должна быть реализована в виде класса `StackBasedQueue`. Интерфейс класса должен включать в себя методы `enqueue()`, `dequeue()`, `__len__()` (остальные аттрибуты должны быть непубличными). При попытке извлечения элемента из пустой очереди должно возбуждаться исключение `Empty`. Конструктор класса должен создавать пустую очередь и не принимать аргументов, кроме `self`.

### Пример применения класса `StackBasedQueue`
```python
>>> queue = StackBasedQueue()
>>> queue.enqueue(1)
>>> queue.enqueue(2)
>>> len(queue)
2
>>> queue.dequeue()
1
>>> len(queue)
1
>>> queue.dequeue()
2
>>> queue.dequeue()
Traceback (most recent call last):
  File "<stdin>", line 1, in <module>
  File "<stdin>", line 16, in dequeue
__main__.Empty
>>> 
```

### Критерии

Правильное решение, выполненное при полном соблюдении требований из условия задачи, оценивается в 1.5 балла.

````python
class Empty(Exception):
    def __init__(self, message=''):
        super().__init__(message)


class StackBasedQueue:
    def __init__(self):
        self._in_stack = []
        self._out_stack = []
        self._size = 0
        
    def __len__(self):
        return self._size
    
    def enqueue(self, v):
        self._in_stack.append(v)
        self._size += 1
        
    def dequeue(self):
        if not self._out_stack:
            while self._in_stack:
                self._out_stack.append(self._in_stack.pop())
        if not self._out_stack:
            raise Empty()
        self._size -= 1
        return self._out_stack.pop()


import random
q = StackBasedQueue()
for _ in range(100):
    if random.randint(0, 1) == 0:
        op = 'enqueue'
    else:
        op = 'dequeue'
    v = random.randint(0, 100)
    if op == 'enqueue':
        print('\nenqueue {}'.format(v))
        q.enqueue(v)
    else:
        print('\nperforming dequeue\ndequeued value: {}'.format(q.dequeue()))
````

# Задача 2. Вечеринка Шелдона

За ужином Леонард сказал Шелдону, что ему уже пора как-то разнообразить свою жизнь. Поэтому Шелдон, поразмыслив, отправил Харрисону Форду приглашение на свой день рождения. 

Шелдон раздал гостям бейджики с номерами от `1` до `N` так, чтобы они показывали порядок прибытия гостей на вечеринку. Для каждой пары номеров `i`, `j` Вам известно знает ли гость под номером `i`, кто такой гость `j` и, наоборот, знает ли `j`-й гость гостя под номером `i`.  (Из того, что `i`-й гость узнал `j`-го, не следует, что `j`-й узнает `i`-го.)

Разумеется, на мероприятии все узнали Харрисона Форда, но мистер Форд не был знаком ни с кем из присутствующих. Определите номер на бейджике Харрисона Форда.

Известно, что актер - единственный, кто всем знаком, и при этом не знает никого из присутствующих.

### Входные данные

В первой строке подаются число гостей `N` и число "узнаваний" `k`, а в следующих `k` строках - `k` пар натуральных числе от `1` до `N`, разделенных пробелом. Пара чисел `1 2` значит, что 1-й гость узнает 2-го гостя.

### Выходные данные

Одно число - номер на бейджике Харрисона Форда.

### Рекомендации

"Узнавания" гостями друг друга можно хранить в виде матрицы смежности, множества пар номеров, словаря, в котором ключами являются пары номеров. Использование списка менее эффективно, если Вы планируете  постоянно обращаться к "узнаваниям".

### Критерии

1. Решение, работающее за линейное время и использующее стек, оценивается в 3 балла.

- Решение, работающее за линейное время, и не использующее стек оуенивается в 2,5 балла.

- Решения, которые функционируют медленней чем за линейное время, оцениваются в 0,5 балла.

````python
import numpy as np

N, k = map(int, input().split())
recognitions = np.zeros([N, N])
for _ in range(k):
    i, j = map(int, input().split())
    recognitions[i-1, j-1] = 1
    
def find_Harrison_Ford(N, k, recognitions):
    guests = list(range(1, N+1))

    while len(guests) > 1:
        g1 = guests.pop()
        g2 = guests.pop()
        if recognitions[g1-1, g2-1]:
            guests.append(g2)
        else:
            guests.append(g1)
    return guests[0]
        
print(find_Harrison_Ford(N, k, recognitions))
````

# Задача 3. Красим дерево

Задано корневое дерево, состоящее из $n$ вершин. Будем считать, что вершины дерева пронумерованы целыми числами от $1$ до $n$. Корнем дерева является вершина с номером $1$.

Каждая вершина дерева имеет определенный цвет, цвет вершины $v$ будем обозначать $c_v$, изначально $c_v = 0$.

В задаче вам требуется раскрасить дерево в заданные цвета с помощью минимального числа действий. За одно действие вы можете выбрать вершину $v$ и цвет $x$, и покрасить все вершины поддерева $v$ (включая $v$) в цвет $x$, то есть для всех вершин $u$ таких, что путь от корня до $u$ содержит вершину $v$, сделать $c_u = x$.

Гарантируется, что все вершины надо покрасить в цвет, отличный от $0$.

Определение корневого дерева можно прочитать по ссылке: https://ru.wikipedia.org/wiki/Дерево_(теория_графов).

### Входные данные
В первой строке находится одно целое число $n$ ($2 \le n \le 10^4$) — число вершин дерева.

В следующей строке следуют $n - 1$ целых чисел $p_2, p_3, \ldots, p_n$ ($1 \le p_i < i$), где $p_i$ обозначает, что существует ребро в графе между вершинами $i$ и $p_i$.

В следующей строке следуют $n$ целых чисел $c_1, c_2, \ldots, c_n$ ($1 \le c_i \le n$), где $c_i$ обозначает желаемый цвет вершины $i$.

Гарантируется, что заданный граф является деревом.

### Выходные данные
На первой и единственной строке выведите одно целое число — минимальное число действий, за которое можно покрасить дерево в желаемые цвета.

# Задача 4. Путешествие

В Семи Королевствах n городов и $n - 1$ дорога, каждая дорога соединяет два города, и возможно достичь каждый город из каждого, передвигаясь по дорогам.

Теон и Яра Грейджой начинают путешествовать на лошади по дорогам из первого города. Из-за тумана они не видят, куда лошадь их везет. Когда лошадь входит в город (в том числе в первый), она направляется в один из городов, соединенных с ним. Но это странная лошадь, поэтому она пойдет только в город, в котором до этого не была. Она идет равновероятно в каждый из таких городов и останавливается, если таких городов нет.

Пусть длина каждой дороги $1$. Путешествие начинается в городе $1$. Чему равно математическое ожидание длины их путешествия? Формально, математическое ожидание — это ожидаемое (то есть среднее) значение. Подробнее можно прочесть по [ссылке](https://ru.wikipedia.org/wiki/Математическое_ожидание).

### Входные данные
Первая строка содержит одно целое число $n$ ($1 \le n \le 100000$) — количество городов.

Далее следует $n - 1$ строк. $i$-я из них содержит два целых числа $u_i$ и $v_i$ ($1 \le u_i$, $v_i \le n$, $u_i \ne v_i$) — номера городов, соединенных $i$-й дорогой.

Гарантируется, что возможно достичь каждый город из каждого, передвигаясь по дорогам.

### Выходные данные
Выведите одно число — математическое ожидание длины путешествия. Путешествие начинается в городе $1$.

Ваш ответ будет считаться правильным, если его абсолютная или относительная точность не превосходит $10^{-6}$.

А именно, если ваш ответ равен $a$, а ответ жюри равен $b$, то ваш ответ будет зачтен, если $\frac{|a - b|}{\max(1, b)} \le 10^{-6}$.

````python
import sys
sys.setrecursionlimit(100002)

n = int(input())

roads_by_cities = tuple([] for _ in range(n))
for _ in range(n-1):
    i, j = map(int, input().split())
    roads_by_cities[i-1].append(j)
    roads_by_cities[j-1].append(i)
    

def expected_path_length(city, prev_city):
    n = 0
    sum_of_expectations = 0
    for c in roads_by_cities[city-1]:
        if c != prev_city:
            n += 1
            sum_of_expectations += 1 + expected_path_length(c, city)
    if n == 0:
        return 0.
    return sum_of_expectations / n


print(expected_path_length(1, -1))
````

### <font color=red>Решение Максима Вовченко</font>

````python
from collections import deque

def expected_value(net_of_cities, n):
    nodes = []
    
    dist = [None] * n
    dist[0] = 0
    
    prob = [0] * n
    prob[0] = 1
    
    prev_nodes = [False] * n
    prev_nodes[0] = True

    current_nodes = deque([0])
    #print('\n')
    
    while current_nodes != deque([]):
        #print('current_nodes: ', current_nodes)
        where = current_nodes.popleft()
        #print('where: ', where)
        
        is_node = True
        
        #print('prev_nodes: ', prev_nodes)
                
        node_count = 0
        for i in net_of_cities[where]:
            if prev_nodes[i] == False:
                node_count += 1
        
        #print('node_count: ', node_count)
        #print('=======')
        
        if node_count > 1:
            prob[where] /= node_count

        for direction in net_of_cities[where]:
            if not prev_nodes[direction]:
                current_nodes.append(direction)
                dist[direction] = dist[where] + 1
                prob[direction] = prob[where]
                prev_nodes[direction] = True
                is_node = False

        if is_node == True:
            nodes.append(where)
        #print('nodes: ', nodes)
        #print('=======')
        
    expected_value = 0
    for i in nodes:
        expected_value += dist[i] * prob[i]
    return expected_value
    


n = int(input())
net_of_cities = [list() for i in range(n)]

for i in range(n-1):
    u, v = map(int, input().split())

    net_of_cities[u-1].append(v-1)
    net_of_cities[v-1].append(u-1)

print(expected_value(net_of_cities, n))
````
