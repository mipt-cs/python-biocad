# Упражнение 4. Медиана

Создайте тип данных `Median`, который позволяет за константное время (сразу) получать медиану набора чисел. Под медианой набора из $N$ чисел понимается `(N // 2)`-е по величине число.  Тип данных `Median` обладает методами:

- `get_median()` для получения медианы,

- `insert()` для добавления нового элемента в набор данных (должен работать за логарифмическое время),

- `remove()` для удаления из набора данных текущей медианы (должен работать за логарифмическое время).

###  Подсказка

Реализуйте класс `Median` с помощью двух двоичных куч `left` и `right`. Куча `left` будет максимум-ориентированной, то есть в ней каждый узел будет больше своих детей (как в определении кучи из alghw12.ipynb). Куча `right` должна быть минимум-ориентированной, и в ней наоборот - каждый узел меньше своих детей. 

У `left` и `right` общий корень, в котором находится медиана набора данных. Куча `left` содержит числа, которые меньше медианы, а `right` - числа, которые больше медианы.

Для работы c кучей `right` отлично подойдет модуль `heapq`, который поддерживает минимум-ориентированные кучи.
