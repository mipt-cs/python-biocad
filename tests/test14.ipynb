# <font color=blue>Задача 1. Папки и файлы, которых не должно быть в директории</font>

Программа получает на вход путь <font color=red face="monospace">head</font> и список имен <font color=red face="monospace">tails</font>, которых не должно быть по пути <font color=red face="monospace">head</font>.

Список <font color=red face="monospace">tails</font> содержит только одноуровневые пути, то есть в <font color=red face="monospace">tails</font> нет директорий, содержащих другие каталоги и файлы.

Распечатайте пути к папкам и файлам, которые нужно удалить.

В первой строке распечатайте **через пробел (не списком)** подлежащие удалению файлы, во второй - подлежащие удалению папки.

Пусть в одном каталоге с данным ноутбуком присутствует папка <font color=green face="monospace">dir</font> со следующей структурой
```
dir
├── dir1
│   ├── dir2
│   ├── dir3
│   ├── file3
│   ├── file4
│   └── file5
├── dir4
│   ├── dir5
│   ├── dir6
│   │   └── file7
│   └── file6
├── file1
└── file2
```

<font color=red>Имена папок не обязательно начинаются со строки **dir**, а имена файлов не обязательно начинаются со строки **file**.</font>

Папку с такой внутренней структурой можно создать с помощью 2-х команд<br>
UNIX
```bash
mkdir -p dir/dir1/dir2 dir/dir1/dir3 dir/dir4/dir5 dir/dir4/dir6
touch dir/file1 dir/file2 dir/dir1/file3 dir/dir1/file4 dir/dir1/file5 dir/dir4/file6 dir/dir4/dir6/file7
```
Windows
```
md dir\dir1\dir2 dir\dir1\dir3 dir\dir4\dir5 dir\dir4\dir6
for %x in (dir\file1 dir\file2 dir\dir1\file3 dir\dir1\file4 dir\dir1\file5 dir\dir4\file6 dir\dir4\dir6\file7) do type nul > %x
```

**Не забудьте проверить, что получилось, с помощью команды**<br>
UNIX
```bash
tree dir
```
WINDOWS
```
tree /F dir
```

### Указания

# <font color=blue>Задача 3. Путешествие</font>

На вход подаются группы имен файлов и папок и натуральные числа, пока не будет введена строка `"STOP"`. Соберите из поданных вход имен путь в файловой системе следующим образом.

1. Имена файлов и папок соединяются в том порядке, в котором они были поданы на вход.

2. Если на вход подано натуральное число $N$, то к пути дописывается $N$ переходов в родительский каталог `".."`.

Распечатайте полученный путь и проверьте, существует ли он.

**Гарантируется, что нет имен папок и файлов, состоящих только из цифр.**

### Указания

1. Разрешается использовать метод `str.split()`, чтобы разбить входные данные по пробелам и `str.isdigit()` для определения, является ли строка натуральным числом. В остальных случай строковые методы не должны применяться.

2. Не используйте срезы, обращение по индексу к элементу строки, конкатенацию.

3. Пути должны собираться с помощью метода `os.path.join()`. 

Пусть в одном каталоге с данным ноутбуком присутствует папка <font color=green face="monospace">dir</font> со следующей структурой
```
dir
├── dir1
│   ├── dir2
│   ├── dir3
│   ├── file3
│   ├── file4
│   └── file5
├── dir4
│   ├── dir5
│   ├── dir6
│   │   └── file7
│   └── file6
├── file1
└── file2
```
