# <font color=blue>Задача 1. Удаление файлов и папок</font>

Удалите из директории все файлы, имена которых содержат поданную на вход строку <font color=red face="Courier" size=3>ban_file</font>, и все папки, имена которых содержат поданную на вход строку <font color=red face="Courier" size=3>ban_dir</font>. 

На вход через пробел подаются путь к директории, строка <font color=red face="Courier" size=3>ban_file</font> и строка <font color=red face="Courier" size=3>ban_dir</font>.

**В примере до удаления файлов и папок подразумевается, что до удаления директория  <font color=green face="Courier" size=3>dir</font> имеет следующую структуру.** 
```
dir
├── hate_python_dir     # directory
│   └── file
├── hate_python_file_1  # file
├── hate_python_file_2  # file
├── love_python_dir     # directory
├── python_hate_file_3  # file
└── python_love_file_1  # file
```

Папку с такой внутренней структурой можно создать с помощью 2-х команд<br>
UNIX
```bash
mkdir -p dir/hate_python_dir dir/love_python_dir
touch dir/hate_python_file_1 dir/hate_python_file_2 dir/python_love_file_1 dir/python_hate_file_3 dir/hate_python_dir/file
```
Windows
```
md dir\hate_python_dir dir\love_python_dir
for %x in (dir\hate_python_file_1 dir\hate_python_file_2 dir\python_love_file_1 dir\python_hate_file_3 dir\hate_python_dir\file) do type NUL > %x
```

**Не забудьте проверить, что получилось, с помощью команды**<br>
UNIX
```bash
tree dir
```
Windows
```
tree /F dir
```
