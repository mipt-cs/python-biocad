### Упражнение 1. Умножение списков

Даны три списка целых числе `A`, `B` и `C`. Элементы списка `D` являются произведениями элементов списков `A`, `B` и `C`, стоящих на соответствующих позициях. Перепишите ниже приведенный код с использванием генераторов списков и `zip`. 

Элементы списков подаются на вход программе через пробел, каждый список в отдельной строке.

Элементы списка `D` череез пробел так же, как выходные данные. 

| <font size=3>Входные данные</font> | <font size=3>Выходные данные</font> |
| :--- | :--- |
| <font size=3>1 1 1<br>1 2 3<br>2 3 4</font> | <font size=3>2 6 12</font> |

````python
A = [int(el) for el in input().split()]
B = [int(el) for el in input().split()]
C = [int(el) for el in input().split()]
D = []
for i in range(len(A)):
    D.append(A[i] * B[i] * C[i])
print(' '.join(D))
````
