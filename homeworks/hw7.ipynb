1. Реализуйте функцию `pow_c()`, вычисляющую натуральные степени $a^N$, $N \in \mathbb{N}$. Вычисление должно производиться простейшим способом в цикле без использования рекурсии.

2. Реализуйте функцию `pow_r()`, для вычисления натуральных степеней с помощью рекурсии.

3. Реализуйте функцию `pow_fast_c()`, вычисляющую натуральные степени числа $a$. Вычисление должно производиться без использования рекурсии. **Оптимизируйте алгоритм с помощью перевода $N$ в двоичную систему счисления.** Пусть требуется вычислить $a^N$, а натуральное число $N$ в двоичной системе счисления имеет вид  $\alpha_k \alpha_{k-1} ... \alpha_0$, где $\left\lbrace \alpha_i \right\rbrace^k_{i=0}$ - цифры в двоичной системе счисления. Получаем 
$$\large {a^N = a^{\sum_{i=0}^k \alpha_i 2^i} = \displaystyle\prod_{i=0}^k \left(a^\left(2^i\right)\right)^{\alpha_i}}.$$
Рассмотрим более пристально выражение $a^\left(2^i\right)$. Легко видеть, что
$$\large{a^\left(2^{i-1}\right) \cdot a^\left(2^{i-1}\right) = a^{2^{i-1} + 2^{i-1}} = a^{2 \cdot 2^{i-1}} = a^\left(2^i\right)}$$
Таким образом, при вычислении  натуральной степени числа $a$ должен производится перевод показателя степени $N$ из десятичной системы счисления в двоичную. Если результат (число $a^N$) собирается в переменной $pow$, то  на итерации цикла вместо добавления разряда в двоичную запись $N$ в $pow$ умножается на $a^{\alpha_i 2^i}$. Кроме того, на каждой итерации с помощью формулы
$$\large{a^\left(2^i\right) = a^\left(2^{i-1}\right) \cdot a^\left(2^{i-1}\right)}$$
пересчитывается $a^\left(2^i\right)$.

4. Реализуйте рекурсивную функцию `pow_fast_r()` для вычисления $a^N$, использующую оптимизацию из пункта 3.

- Определите скорость работы алгоритмов. Пример того, как это можно сделать, есть в следующей клетке.

- Как Вы считаете, возможно ли повысить эффективность алгоритма из пункта 3, если переводить $N$ в систему счисления с большим основанием? Ответ обоснуйте.

````python
import math
import time
start = time.clock()
N = 1000
for _ in range(N):
    math.factorial(N)
print("Среднее время вычисления факториала:", (time.clock() - start) / N)
````
