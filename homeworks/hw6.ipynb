$$S(n) = \sqrt{2 \pi n} \left(\frac{n}{e}\right)^n $$

$$n! = S(n) \left( 1 + O\left(\frac{1}{n}\right)\right)$$

Для каждого $n$ определите величину $C_n$ - константу в $O(\frac{1}{n})$.

$$C_n = \frac{n \left(n! - S(n)\right)}{S(n)}$$

>Запись
$$f(x) = O\left(g(x)\right),~ x \to x_0,$$
где $x_0 \in \mathbb{R} \cup \infty$ означает, что в некоторой проколотой окрестности $\dot{U}_\delta(x_0)$ точки $x_0$
определены функции $f$, $g$, $\varphi$ такие, что 
$$f(x) = g(x) \varphi(x),$$
где $\varphi(x)$ - функция, ограниченная на $\dot{U}_\delta(x_0)$. *(Тер-Крикоров А.М., Шабунин М.И. "Курс математического анализа" - 3 изд., 2001 - стр. 119)*<br>
Другими словами
$$f(x) \stackrel{x \to x_0}{=} O(g(x)) \iff \exists C > 0: \left|f(x)\right| \leq C \left|g(x)\right|~ \forall x \in \dot{U}_\delta(x_0)$$

````python
import math
from decimal import Decimal
print(math.pi)  # Число пи
print(math.e)  # Число e
d = Decimal(3)  # Создаем Decimal из wtkjuj xbckf
print(d * 4)  # OK так как 4 - int
print(d * Decimal(4.))  # OK, т. к. float был заменен на Decimal
print(d * 4.)  # Error
````
